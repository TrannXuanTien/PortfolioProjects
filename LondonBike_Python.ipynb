```python
#Rename columns for easier understanding
new_columns_name = {
     'timestamp':'time',
     'cnt':'count',
     't1':'temp_actual_C',
     't2':'temp_feel_C',
     'hum':'humidity_percent',
     'wind_speed':'wind_speed',
     'weather_code':'weather',
     'is_holiday':'is_holiday',
     'is_weekend':'is_weekend',
     'season':'season'
 }
london_bike.rename(new_columns_name,axis=1,inplace=True)
london_bike
```

```python
# Changing numerical values of weather and season to written values

# Creating dictionaries of written values to map to the columns
holiday_written = {'0.0':'No',
                  '1.0':'Yes'}

weekend_written = {'0.0':'No',
                  '1.0':'Yes'}

season_written = {'0.0':'Spring',
                 '1.0':'Summer',
                 '2.0':'Fall',
                 '3.0':'Winter'}


weather_written = {'1.0':'Clear',
                  '2.0':'Scattered Clouds',
                  '3.0':'Broken Clouds',
                  '4.0':'Cloudy',
                  '7.0':'Rain',
                  '10.0':'Rain With Thunderstorm',
                  '26.0':'Snowfall',
                  '94.0':'Freezing Fog'}

# Changing type of is_holiday to string
london_bike.is_holiday = london_bike.is_holiday.astype('str')
# Mapping values from created dictionary to is_holiday column
london_bike.is_holiday = london_bike.is_holiday.map(holiday_written)

# Changing type of is_weekend to string
london_bike.is_weekend = london_bike.is_weekend.astype('str')
# Mapping values from created dictionary to is_holiday column
london_bike.is_weekend = london_bike.is_weekend.map(weekend_written)

# Changing type of season to string
london_bike.season = london_bike.season.astype('str')
# Mapping values from created dictionary to season columns 
london_bike.season = london_bike.season.map(season_written)

# Changing type of weather to string
london_bike.weather = london_bike.weather.astype('str')
# Mapping values from created dictionary to weather columns
london_bike.weather = london_bike.weather.map(weather_written)
```

```python
#Double check to see whether the values is changed or not
london_bike
```
